### Feature Engineering

#### Previous searches

I'm going to extract different features from the 'previous_search' field to analyse it given a search_id. I'm going to focus only in:
- if has made a conversion (each one).
- if has searched the same trip for same origin - destination (how many yes?).
- how many _OW_ or _RT_ trips has searched (each one).
- average time for _RT_ trips.
- how many previous searches were?

For those searches that doesn't have any previous_search information, all of the features listed will output 0.

```python
def made_conversion_previous(previous_search, travel_search):
    x_dict = eval(previous_search)
    conversion_yes_p, conversion_not_p = 0, 0
    
    ##conversion_type
    if x_dict['conversion']:
        conversion_yes_p = int(np.sum([1 if x == True else 0 for x in x_dict['conversion']]))
        conversion_not_p = int(np.sum([1 if x == False else 0 for x in x_dict['conversion']]))
    
    ##has searched same trip
    previous_same_search = 0
    if x_dict['origin_destination']:
        previous_same_search = np.sum([1 if x == travel_search else 0 for x in x_dict['origin_destination']])
        
    ##travel_type
    previous_rt, previous_ow = 0, 0
    if x_dict['travel_type']:
        previous_rt = int(np.sum([1 if x == 'RT' else 0 for x in x_dict['travel_type']]))
        previous_ow = int(np.sum([1 if x == 'OW' else 0 for x in x_dict['travel_type']]))
        
    ##average time for RT types
    avg_rt_days_previous = 0
    if x_dict['travel_type']:
        list_rt_only = [travel_date for travel_type, travel_date \
                                        in zip(x_dict['travel_type'], x_dict['travel_dates']) \
                                        if travel_type == 'RT']
        if list_rt_only:
            avg_rt_days_previous = np.round(np.mean([(datetime.strptime(x.split('_')[1],'%Y-%m-%d') - \
                                        datetime.strptime(x.split('_')[0], '%Y-%m-%d')).days \
                                             for x in list_rt_only]), 2)
    
    ##previous searches
    ##len(previous_searches)
    

    return(pd.Series(dict(
        conversion_yes_p = conversion_yes_p,
        conversion_not_p = conversion_not_p,
        previous_searches = len(x_dict['conversion']),
        previous_same_search = previous_same_search,
        previous_rt = previous_rt,
        previous_ow = previous_ow,
        avg_rt_days_previous = avg_rt_days_previous
    )))

df_training = pd.concat([df_train, 
                         df_train[['previous_searches', 'travel_search']]\
                            .apply(lambda df: made_conversion_previous(df.previous_searches, df.travel_search), axis=1)],
                         axis = 1)
df_training.head(5)
```

### Training model

In this part of the code, I want to train a model and look how it is bahaving with known labeled data in order to predict the outcome of test dataset. I will create a basic model with few hyperparameter tunning options and then choose the best of that model. The best model is going to predict the labeled test dataset.

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn import  preprocessing
from sklearn import metrics
from sklearn.metrics import roc_curve, precision_recall_curve, auc, make_scorer, recall_score,\
                            accuracy_score, precision_score, confusion_matrix, f1_score
from sklearn.model_selection import train_test_split, StratifiedKFold, GridSearchCV
from sklearn.preprocessing import LabelBinarizer

##now df_training_set
objective = 'conversion'
ids = ['search_id']

lb = LabelBinarizer()
df_training_set[objective] = lb.fit_transform(df_training_set[objective].values)
targets = df_training_set[objective]


y = df_training_set[objective]
X = df_training_set[list(set(list(df_training_set.columns)) - set(list([objective]) + list(ids)))]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, shuffle = True, stratify = targets)

print('data entrenamiento')
print('... ')
print('Número de cols: ' + str(len(X_train.columns)))
print('Número de observaciones: ' + str(len(X_train)))
print('...')


print('data test')
print('... ')
print('Número de cols: ' + str(len(X_test.columns)))
print('Número de observaciones: ' + str(len(X_test)))
print('...')
```

```python
print('y_train class distribution')
print(y_train.value_counts(normalize=True))
print('y_test class distribution')
print(y_test.value_counts(normalize=True))
```

```python
clf = RandomForestClassifier(n_estimators=100)
clf.fit(X_train, y_train)

y_pred_base = clf.predict(X_test)

print("F1 score Base model:",metrics.f1_score(y_test, y_pred_base))
```

```python
feature_importances = pd.DataFrame(clf.feature_importances_,
                                   index = X_train.columns,
                                   columns=['importance'])\
    .sort_values('importance', ascending=False)

print('DecisionTreesClassifier')
print()
feature_importances
```

```python
clf2 = RandomForestClassifier(n_jobs=-1)

param_grid = {
    'min_samples_split': [3, 5, 10], 
    'n_estimators' : [200, 300, 400],
    'max_depth': [3, 5, 15, 25],
    'max_features': [10, 15, 20, 25]
}

scorers = {
    'precision_score': make_scorer(precision_score),
    'recall_score': make_scorer(recall_score),
    'f1_score': make_scorer(f1_score)
}
```

```python
def grid_search_wrapper(refit_score='f1_score'):
    """
    fits a GridSearchCV classifier using refit_score for optimization
    prints classifier performance metrics
    """
#     skf = StratifiedKFold(n_splits=3)
    grid_search = GridSearchCV(clf2, param_grid, scoring=scorers, refit=refit_score,
                               cv=3, return_train_score=True, n_jobs=-1, verbose=2)
    grid_search.fit(X_train.values, y_train.values)

    # make the predictions
    y_pred_model = grid_search.predict(X_test.values)

    print('Best params for {}'.format(refit_score))
    print(grid_search.best_params_)

    # confusion matrix on the test data.
    print('\nConfusion matrix of Random Forest optimized for {} on the test data:'.format(refit_score))
    print(pd.DataFrame(confusion_matrix(y_test, y_pred_model),
                 columns=['pred_neg', 'pred_pos'], index=['neg', 'pos']))
    return grid_search

grid_search_clf = grid_search_wrapper(refit_score='f1_score')
```

```python
print("F1 score best model fit:",metrics.f1_score(y_test, grid_search_clf.best_estimator_.predict(X_test.values)))
```

```python
print('Table best scores: ');print()
results = pd.DataFrame(grid_search_clf.cv_results_)
results = results.sort_values(by='mean_test_f1_score', ascending=False)
results[['mean_test_precision_score', 'mean_test_f1_score',
         'param_max_depth', 'param_max_features', 
         'param_min_samples_split', 'param_n_estimators']].round(3).head(5)
```

```python
y_scores = grid_search_clf.predict_proba(X_test)[:, 1]
p, r, thresholds = precision_recall_curve(y_test, y_scores)
```

```python
def plot_precision_recall_vs_threshold(precisions, recalls, thresholds):
    """
    Modified from:
    Hands-On Machine learning with Scikit-Learn
    and TensorFlow; p.89
    """
    plt.figure(figsize=(8, 8))
    plt.title("Precision and Recall Scores as a function of the decision threshold")
    plt.plot(thresholds, precisions[:-1], "b--", label="Precision")
    plt.plot(thresholds, recalls[:-1], "g-", label="Recall")
    plt.ylabel("Score")
    plt.xlabel("Decision Threshold")
    plt.legend(loc='best')
```

```python
print('Precision Vs. Recall plot: ');print()
plot_precision_recall_vs_threshold(p, r, thresholds)
```

```python

def plot_roc_curve(fpr, tpr, label=None):
    """
    The ROC curve, modified from 
    Hands-On Machine learning with Scikit-Learn and TensorFlow; p.91
    """
    plt.figure(figsize=(8,8))
    plt.title('ROC Curve')
    plt.plot(fpr, tpr, linewidth=2, label=label)
    plt.plot([0, 1], [0, 1], 'k--')
    plt.axis([-0.005, 1, 0, 1.005])
    plt.xticks(np.arange(0,1, 0.05), rotation=90)
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate (Recall)")
    plt.legend(loc='best')
    
fpr, tpr, auc_thresholds = roc_curve(y_test, y_scores)
print(auc(fpr, tpr)) # AUC of ROC
plot_roc_curve(fpr, tpr, 'roc')
```
